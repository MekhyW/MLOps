```python
import pandas as pd
import time
import matplotlib.pyplot as plt
```

```python
times_csv = []
times_json = []
times_parquet = []
```

```python
for i in range(1000):
    start = time.time()
    df = pd.read_csv("ex_data.csv")
    end = time.time()
    times_csv.append(end-start)
```

```python
for i in range(1000):
    start = time.time()
    df = pd.read_json("ex_data.json")
    end = time.time()
    times_json.append(end-start)
```

```python
for i in range(1000):
    start = time.time()
    df = pd.read_parquet("ex_data.parquet")
    end = time.time()
    times_parquet.append(end-start)
```

```python
plt.boxplot([times_csv, times_json, times_parquet])
plt.xticks([1, 2, 3], ['CSV', 'JSON', 'Parquet'])
plt.ylabel("Time to read file (s)")
plt.show()
```

```python
df = pd.DataFrame({
    "CSV": times_csv,
    "JSON": times_json,
    "Parquet": times_parquet
})
print(df.describe())
```
